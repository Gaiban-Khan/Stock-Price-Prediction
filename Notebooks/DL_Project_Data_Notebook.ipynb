```python
import pandas as pd
import glob
```

```python
path = "/content/NSE-TATAMOTORS"
files = glob.glob(path + "/*.csv")
```

```python
df = pd.concat(pd.read_csv(f) for f in files)
```

```python
df.head()
```

```python
df.drop(columns=["Symbol", "Series", "No. of Trades", "Prev Close", "Average Price"], inplace=True)
df.rename(columns = {"Open Price" : "Open", "High Price" : "High", "Low Price" : "Low", "Last Price" : "Last", "Close Price" : "Close"}, inplace=True)
```

```python
df.head()
```

```python
df.to_csv("NSE-TATAMOTORS.csv", index=False)
```
